 1 Calculate the overall attrition percentage in the company

    use the Attrition and age column.

    find the attrition percentage in age column

```python
#1 Calculate the overall attrition percentage in the company
age_counts = a.groupby('Attrition')['Age'].count()
attrition_percentage = (age_counts.get('Yes', 0) / age_counts.sum()) * 100
print("Overall Attrition Percentage:", round(attrition_percentage, 2), "%")
```

```python
import matplotlib.pyplot as plt
age_counts = a.groupby('Attrition')['Age'].count()
# Step 2: Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.2f%%', startangle=90, colors=['skyblue', 'salmon'])
plt.title("Overall Attrition Distribution")
plt.axis('equal')  # Keeps the pie circular
plt.show()
```

A large majority (83.88%) of employees have stayed, indicating overall stability in the workforce.

The 16.12% attrition rate, while not extreme, is meaningful enough to explore further — especially across departments, roles, or overtime patterns.

 2.Identify the top 3 job roles with the highest attrition rates.

            JobRole → to group employees by their role

            Attrition → to count how many left vs stayed

            then find the attrition of jobrole in age column

            find the percentage of top 3 jobroles

```python
#2.Identify the top 3 job roles with the highest attrition rates.
role_counts = a.groupby(['JobRole', 'Attrition'])["Age"].count().unstack(fill_value=0)
role_counts
```

```python
role_counts['AttritionRate'] = (role_counts['Yes'] / (role_counts['Yes'] + role_counts['No'])) * 100
top_roles = role_counts.sort_values(by='AttritionRate', ascending=False).head(3)
print(top_roles[['AttritionRate']])
```

```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.barplot(x=top_roles.index, y=top_roles['AttritionRate'], palette='Reds')
plt.title("Top 3 Job Roles by Attrition Rate")
plt.xlabel("Job Role")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=30)
plt.show()
```

Sales Representatives face the highest turnover, with an attrition rate around 40%, suggesting this role may involve high pressure, limited growth, or unmet expectations.

Laboratory Technicians and HR roles also show elevated attrition, both near 25%, indicating potential issues with job satisfaction, career progression, or workload in these positions.

```python
plt.figure(figsize=(7, 5))
sns.barplot(x=agegroup_counts.index, y=agegroup_counts['AttritionRate'], palette='crest')
plt.title("Attrition Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Attrition Rate (%)")
plt.ylim(0, 30)
plt.show()
```

Employees under 30 have the highest attrition rate (~25%), indicating younger professionals may be more likely to switch jobs for better opportunities, growth, or flexibility.

Attrition decreases with age, with the 30–40 group around 13% and the >40 group just above 10%, suggesting older employees may value stability or have stronger organizational ties.
